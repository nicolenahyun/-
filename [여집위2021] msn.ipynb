```python
def get_media(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    try:
        media = soup.find('span', 'partner').text.strip()
        if media == None:
            media = 'N/A'
    except:
        media = 'N/A'
    return media


# N/A인 부분부터 다시 돌리기

for index in msn.loc[7042:].index:
    url = msn.loc[index, 'FULL_URL']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    
    media = get_media(url)

    if index % 5 == 0:
        print(url, media)
        time.sleep(0.5)

    msn.loc[index, 'MEDIA'] = media
```

```python
#한 곳까지 저장
writer = pd.ExcelWriter('여집위2021_msn_매체분류.xlsx', engine='xlsxwriter', options={'strings_to_urls': False})
msn.to_excel(writer, index=False)
writer.close()
```
