```python
oid = pd.read_excel('네이버_언론사별oid.xlsx', header=2)
print(len(oid))
oid.head()
```

```python
drop_index = oid[oid['제호'] == '언론사별 뉴스'].index
oid.drop(drop_index, axis='index', inplace=True)
oid
```

```python
oid.drop_duplicates(inplace=True)
oid.reset_index(drop=True, inplace=True)
oid
```

```python
oid.to_excel('네이버_언론사별oid_중복제외.xlsx', index=False)
```

```python
oid['oid_pad'] = oid['oid'].astype('str').str.pad(width=3, side='left', fillchar='0')
oid.head()
```

```python
oid_dict = {}
for index in oid.index:
    oid_pad = oid.loc[index, 'oid_pad']
    medium = oid.loc[index, '제호']
    oid_dict[oid_pad] = medium

print(len(oid_dict))
```

```python
url = naver['FULL_URL'][3]
re.findall(r'oid=(\d+)&', url)
```

```python
for index in naver.index:
    url = naver.loc[index, 'FULL_URL']
    try:
        oid = re.findall(r'oid=(\d+)&', url)[0]
        if oid in oid_dict:
            media = oid_dict[oid]
        else:
            media = 'N/A'
    except:
        media = 'N/A'
    naver.loc[index, 'MEDIA'] = media
```

```python
def naver_get_media(name):
    df1 = pd.read_excel('여집위2021_매체명분류_팀원별분배.xlsx', sheet_name=name)
    
    print(df1['MEDIA'].isna().sum())
    for index in df1[df1['MEDIA'].isna()].index:
        url = df1.loc[index, 'FULL_URL']
        try:
            oid = re.findall(r'article/(\d+)/', url)[0]
            if oid in oid_dict:
                media = oid_dict[oid]
                df1.loc[index, 'MEDIA'] = media
        except:
            continue
    
    print(df1['MEDIA'].isna().sum())
    return df1
```

```python
writer = pd.ExcelWriter('여집위2021_매체명분류_네이버추가분류.xlsx', options={'strings_to_urls': False})

naver_get_media('이빛나').to_excel(writer, sheet_name='이빛나', index=False)
naver_get_media('노용정').to_excel(writer, sheet_name='노용정', index=False)
naver_get_media('안민선').to_excel(writer, sheet_name='안민선', index=False)

writer.save()
```
