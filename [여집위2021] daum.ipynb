```python
# 매체명을 scraping해오는 코드를 미리 함수로 작성
def get_media(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    try:
        img = soup.find('em', 'info_cp').find('img')
        media = img.get('alt').strip()
        if media == None:
            media = 'N/A'
    except:
        try:
            media = soup.find('span', 'source').text.strip()
            if media == None:
                media = 'N/A'
        except:
            media = 'N/A'
    return media


for index in daum.index:
    url = daum.loc[index, 'FULL_URL']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    
    if 'cartoon' in url:
        media = 'cartoon'
    elif 'gallery' in url:
        media = 'gallery'
    else:
        if url[-3:] == 'f=o':
            clean_url = url[:-4] 
            media = get_media(clean_url)
        else: 
            media = get_media(url)

    if index % 5 == 0:
        print(url, media)
        time.sleep(0.5)

    daum.loc[index, 'MEDIA'] = media
```

### 추가로 gallery 수집

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
import time
from bs4 import BeautifulSoup


options = Options()
options.add_argument("--headless")  ## headless 모드로 설정
options.add_argument("window-size=1920,1080") ## 브라우저 창 크기 설정(1920x1080) - 너무 크기가 작으면 로딩이 잘 안될 수 있다
driver = webdriver.Chrome(options=options)


for index in daum[daum['MEDIA'] == 'gallery'].index:
    url = daum.loc[index, 'FULL_URL']
    driver.get(url)
    time.sleep(1)  # 잠시 쉬며 로딩될 수 있도록. 
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    media_span = soup.find('span', 'info_cp')

    if media_span is None:  # 잘 못잡아오는 경우는 로딩에 시간이 걸려서라고 판단, 새로고침 & 3초 쉬기를 추가
        driver.refresh() # 새로고침
        time.sleep(3) 
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        media_span = soup.find('span', 'info_cp')
        
    try: media = media_span.text
    except: media = 'N/A'
    

    if index % 5 == 0:
        print(url, media)
        
    daum.loc[index, 'Gallery'] = media
        
driver.close()
```

```python
writer = pd.ExcelWriter('여집위2021_daum_매체분류.xlsx', engine='xlsxwriter', options={'strings_to_urls': False})
daum.to_excel(writer, index=False)
writer.close()
```

## daum finance 수집

```python
# url 목록 가져오기
f = open('daum_finance_url.txt', 'r', encoding='utf-8')
url_list = [line.strip() for line in f.readlines()]
f.close()
print(len(url_list), url_list[0])
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
import time
from bs4 import BeautifulSoup


options = Options()
options.add_argument("--headless")  ## headless 모드로 설정
options.add_argument("window-size=1920,1080") ## 브라우저 창 크기 설정(1920x1080) - 너무 크기가 작으면 로딩이 잘 안될 수 있다
driver = webdriver.Chrome(options=options)


f = open('daum_finance.txt', 'w', encoding='utf-8')

for url in url_list:
    driver.get(url)
    time.sleep(1)  # 잠시 쉬며 로딩될 수 있도록. 
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    media_div = soup.find('div', 'titView')

    if media_div is None:  # 잘 못잡아오는 경우는 로딩에 시간이 걸려서라고 판단, 새로고침 & 3초 쉬기를 추가
        driver.refresh() # 새로고침
        time.sleep(3) 
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        media_div = soup.find('div', 'titView')
        
    try: media = media_div.em.text
    except: media = 'N/A'
    

    print(url, media)
    f.write(url + '\t' + media + '\n')
        
driver.close()
f.close()
```
